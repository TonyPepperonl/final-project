```python
import pandas as pd
df=pd.read_csv("StudentPerformanceFactors.csv")
df
```

Cleansing null values

```python
df.isnull().sum()
```

```python
df['Teacher_Quality'] = df['Teacher_Quality'].fillna(df['Teacher_Quality'].mode()[0]) 
df['Parental_Education_Level'] = df['Parental_Education_Level'].fillna(df['Parental_Education_Level'].mode()[0]) 
df['Distance_from_Home'] = df['Distance_from_Home'].fillna(df['Distance_from_Home'].mode()[0]) 
```

```python
df.isnull().sum()
```

Dropping irrelevant columns

```python
df = df.drop(columns=['Gender', 'Learning_Disabilities','Distance_from_Home', 'Previous_Scores', 'Peer_Influence'])
```

Renaming columns

```python
df = df.rename(columns={'Hours_Studied': 'Hours Studied', 'Attendance': 'Attendance','Parental_Involvement': 'Parental Involvement', 'Access_to_Resources': 'Resource Access','Extracurricular_Activities': 'Extracurricular Activities', 'Sleep_Hours': 'Sleep Hours','Motivation_Level': 'Motivation Level', 'Internet_Access': 'Internet Access','Tutoring_Sessions': 'Tutoring Sessions', 'Family_Income': 'Family Income', 'Teacher_Quality': 'Teacher Quality', 'School_Type': 'School Type', 'Physical_Activity': 'Physical Activity', 'Parental_Education_Level': 'Parental Education Level', 'Exam_Score': 'Exam Score'})
```

```python
df.columns
```

```python
df.head(10)
```

Splitting the data based on data types and One-Hot encoding

```python
df.dtypes
```

```python
cat_col = ['Parental Involvement','Resource Access', 'Motivation Level', 'Family Income', 'Teacher Quality', "Parental Education Level"]
cat_df = df[cat_col]
df = pd.get_dummies(df.drop(*[cat_col], axis=1), drop_first=True)
df.head()
```

```python
for col in cat_df:
    if cat_df[col].str.contains("Low").any():
        levels = {"Low": 0, "Medium": 1, "High": 2}
    else:
        levels = {"High School": 0, "College": 1, "Postgraduate": 2}

    cat_df[col] = cat_df[col].map(levels)

cat_df.head()
```

```python
df = pd.concat([df, cat_df], axis=1)
df.astype(int)
df.head()
```

```python
clean_df = df 
clean_df
```

```python
clean_df.to_csv("cleaned_data.csv", index=False)  
```
